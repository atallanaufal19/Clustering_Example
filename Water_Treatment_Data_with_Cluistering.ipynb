# **Pemodelan**

```python
data_uji = data.values
M = data_uji
```

```python
K = 3 #jumlah cluster
n_iterasi = 50 #jumlah batas maksimal iterasi
```

```python
#pemodelan dengan Kmeans

def kmeans(X, k):
  pemb = 1
  cluster = np.zeros(X.shape[0])#inisisasi cluster
  centroids = data.sample(n=k).values #random centroid
  iterasi = 1
  j = 1
  while pemb and j <= n_iterasi:
    for i, row in enumerate(X):
      jarak_min = float('inf')
      # Eucledian Distance
      for index, centroid in enumerate(centroids):
        eucledianD = np.sqrt((centroid[0]-row[0])**2 + (centroid[1] - row[1])**2)
        # Mengembalikan centroid terdekat
        if jarak_min > eucledianD:
          jarak_min = eucledianD
          cluster[i] = index
    centroid_baru = pd.DataFrame(X).groupby(by=cluster).mean().values
    # Akhiri apabila centroid lama sama dengan centroid baru
    if np.count_nonzero(centroids-centroid_baru) == 0:
      pemb = 0
    # Apabila tidak sama maka centroid baru jadi centroid lama dan akan perulangan lagi
    else:
      centroids = centroid_baru
    #Lihat banyaknya iterasi 
    print("Iterasi ke:", iterasi)
    iterasi = iterasi + 1
    j = j + 1
  return centroids, cluster
```

```python
#menjalankan fungsi Kmeans
centroids, cluster = kmeans(M, K)
```

# **Evaluasi**

```python
#mencari Cluster optimal dengan elbow method

from sklearn.cluster import KMeans

km_scores= []
print(data_uji.shape)
for i in range(1,11):
    km = KMeans(n_clusters=i, random_state=0).fit(data_uji)
    preds = km.predict(data_uji)
    
    print("Skor dari cluster {}: {}".format(i,km.score(data_uji)))
    km_scores.append(-km.score(data_uji))
```

```python
plt.figure(figsize=(12,5))
plt.title("ELBOW METHOD\n",fontsize=18)
plt.plot([i for i in range(1,11)],km_scores, marker = 'o')
plt.grid(True)
plt.xlabel("Banyaknya cluster",fontsize=15)
plt.ylabel("Nilai K-Means ",fontsize=16)
plt.xticks([i for i in range(1,11)],fontsize=15)
plt.yticks(fontsize=16)
plt.savefig('the_elbow_method.png')
plt.show()
```
